```python
import joblib
import pandas as pd
import pyarrow.parquet as pq

# Load the saved model
model = joblib.load("xgb_churn_model.pkl")

# Load or prepare new data for prediction (ensure the same preprocessing is applied)
# Example: Loading a new dataset
new_data = pd.read_parquet(r"C:\Users\utkub\OneDrive\Masaüstü\Code\Capstone\processed.parquet", engine="pyarrow")  # Change this to your actual data source

# Ensure the new data has the same feature columns as the original training data
X_new = new_data.drop(columns=['churn'], errors='ignore')  # Drop churn if it exists

# Make predictions
predictions = model.predict(X_new)
prediction_probabilities = model.predict_proba(X_new)[:, 1]  # Probability of churn

# Add predictions to the dataframe
new_data["churn_prediction"] = predictions
new_data["churn_probability"] = prediction_probabilities

# Save or display results
new_data.to_csv("customer_predictions.csv", index=False)  # Save predictions to CSV
print(new_data.head())  # Display first few rows
```

```python
import pandas as pd
import glob

# Dosya yolu kalıbı (regex uyumlu glob deseni)
file_pattern = "../../../tt_data/capstone.*.jsonl.gz"

# Tek seferde tüm dosyaları oku
combined_df = pd.concat(
    (pd.read_json(file, lines=True, compression='gzip') for file in glob.glob(file_pattern)),
    ignore_index=True
)

# Satır sayısını yazdır
print("Toplam Satır Sayısı:", combined_df.shape[0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Grafik ayarları
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 1️⃣ Numerik ve kategorik sütunlar
numerical_columns = [
    'age', 'tenure', 'avg_call_duration', 'data_usage', 'roaming_usage',
    'monthly_charge', 'overdue_payments', 'avg_top_up_count',
    'call_drops', 'customer_support_calls', 'satisfaction_score'
]

categorical_columns = [
    'id','service_type', 'auto_payment', 'İzleGo', 'RitimGo',
    'CüzdanX', 'HızlıPazar', 'Konuşalım',"churn"
]

# 2️⃣ Klasör oluşturma
output_dir = "plots"
numerical_dir = os.path.join(output_dir, "numerical")
categorical_dir = os.path.join(output_dir, "categorical")

os.makedirs(numerical_dir, exist_ok=True)
os.makedirs(categorical_dir, exist_ok=True)

# 3️⃣ Numerik değişkenler için histogram
for col in numerical_columns:
    plt.figure()
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'{col} - Histogram')
    plt.xlabel(col)
    plt.ylabel('Frekans')
    plt.grid(True)
    
    # Dosya kaydetme
    plt.savefig(os.path.join(numerical_dir, f'{col}_histogram.png'))
    plt.close()
```

```python

# 4️⃣ Kategorik değişkenler için frekans grafiği
for col in categorical_columns:
    plt.figure()
    sns.countplot(x=df[col])
    plt.title(f'{col} - Frekans Grafiği')
    plt.xlabel(col)
    plt.ylabel('Frekans')
    plt.grid(True)
    
    # Dosya kaydetme
    plt.savefig(os.path.join(categorical_dir, f'{col}_frequency.png'))
    plt.close()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count

# Spark oturumunu başlat
spark = SparkSession.builder.appName("ReadIDColumn").getOrCreate()

# Dosya desenini oku
file_pattern = "../../../tt_data/capstone.*.jsonl.gz"

# JSONL dosyalarını oku
df = spark.read.json(file_pattern)

# NA değer sayısını bul
na_count = df.filter(col("id").isNull()).count()
print(f"NA Değer Sayısı: {na_count}")

# 1'den fazla tekrar eden değerleri bul
duplicate_ids = df.groupBy("id").agg(count("id").alias("count")).filter(col("count") > 1)

# Sonuçları göster
if duplicate_ids.count() > 0:
    print("Tekrar Eden Değerler:")
    duplicate_ids.show()
else:
    print("Tekrar eden değer bulunamadı.")
```

```python
from pyspark.sql.functions import when, col, lit, array_contains

# Uygulama listesi
app_list = ['İzleGo', 'RitimGo', 'CüzdanX', 'HızlıPazar', 'Konuşalım']

# Her uygulama için yeni sütun oluştur
for app in app_list:
    df = df.withColumn(app, when(array_contains(col("apps"), app), lit(1)).otherwise(lit(0)))

# Sonucu göster
df.show(truncate=False)
```

```python
from pyspark.sql.functions import expr

# SQL tablosu olarak kaydet
df.createOrReplaceTempView("app_data")

# 1️⃣ Toplam müşteri sayısını hesapla
total_users_query = "SELECT COUNT(DISTINCT id) AS total_users FROM app_data"
total_users_df = spark.sql(total_users_query)
total_users = total_users_df.collect()[0]["total_users"]

# 2️⃣ En az bir uygulama kullanan müşteri sayısını hesapla
active_users_query = """
SELECT COUNT(DISTINCT id) AS active_users 
FROM app_data 
WHERE array_contains(apps, 'İzleGo') 
   OR array_contains(apps, 'RitimGo') 
   OR array_contains(apps, 'CüzdanX') 
   OR array_contains(apps, 'HızlıPazar') 
   OR array_contains(apps, 'Konuşalım')
"""
active_users_df = spark.sql(active_users_query)
active_users = active_users_df.collect()[0]["active_users"]

# 3️⃣ Uygulama bazında kullanım oranlarını hesapla
query = f"""
SELECT
    app_name,
    user_count,
    ROUND((user_count / {total_users}) * 100, 2) AS percentage_of_total_users,
    ROUND((user_count / {active_users}) * 100, 2) AS percentage_of_active_users
FROM (
    SELECT 'İzleGo' AS app_name, COUNT(DISTINCT id) AS user_count FROM app_data WHERE array_contains(apps, 'İzleGo')
    UNION ALL
    SELECT 'RitimGo' AS app_name, COUNT(DISTINCT id) AS user_count FROM app_data WHERE array_contains(apps, 'RitimGo')
    UNION ALL
    SELECT 'CüzdanX' AS app_name, COUNT(DISTINCT id) AS user_count FROM app_data WHERE array_contains(apps, 'CüzdanX')
    UNION ALL
    SELECT 'HızlıPazar' AS app_name, COUNT(DISTINCT id) AS user_count FROM app_data WHERE array_contains(apps, 'HızlıPazar')
    UNION ALL
    SELECT 'Konuşalım' AS app_name, COUNT(DISTINCT id) AS user_count FROM app_data WHERE array_contains(apps, 'Konuşalım')
)
"""

result_df = spark.sql(query)

# 4️⃣ En az bir uygulama kullananların oranını ekle
final_query = f"""
SELECT 
    *, 
    (SELECT ROUND(({active_users} / {total_users}) * 100, 2)) AS active_users_percentage
FROM ({query})
"""

final_result_df = spark.sql(final_query)

# Sonucu ekrana yazdır
final_result_df.show()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.sql.functions import col, when

# Kullanılmayacak sütunları filtrele
exclude_cols = ["id", "apps"]
feature_cols = [c for c in df.columns if c not in exclude_cols]

# Eksik verileri doldurmak için her bir sütun üzerinde işlem yap
for target_col in feature_cols:
    
    # Mevcut sütun dışındaki tüm sütunları feature olarak belirle
    input_features = [c for c in feature_cols if c != target_col]

    # Özellik vektörü oluştur
    assembler = VectorAssembler(inputCols=input_features, outputCol="features")
    df_transformed = assembler.transform(df)

    # Eksik olmayan satırları model eğitimi için kullan
    df_train = df_transformed.filter(col(target_col).isNotNull())
    df_test = df_transformed.filter(col(target_col).isNull())

    # Modeli eğitmek için RandomForestRegressor kullan
    rf = RandomForestRegressor(featuresCol="features", labelCol=target_col)
    model = rf.fit(df_train)

    # Eksik değerleri tahmin et
    predictions = model.transform(df_test).select("id", "prediction")

    # Orijinal df'ye tahminleri ekleyerek eksik değerleri doldur
    df = df.join(predictions, "id", "left_outer").withColumn(
        target_col, when(col(target_col).isNull(), col("prediction")).otherwise(col(target_col))
    ).drop("prediction")

# Sonucu göster
df.show(truncate=False)
```
